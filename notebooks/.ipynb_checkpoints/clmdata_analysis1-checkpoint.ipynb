```python
import pandas as pd

# Load the larger claims dataset
claims_df_large = pd.read_csv('../data/claims_data_large.csv')

# Display the first few rows of the dataset
claims_df_large.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the seaborn style for better aesthetics
sns.set(style='whitegrid')

# Plot distribution of claim amounts
plt.figure(figsize=(10, 6))
sns.histplot(claims_df_large['ClaimAmount'], bins=20, kde=True)
plt.title('Distribution of Claim Amounts')
plt.xlabel('Claim Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Group by ClaimType and calculate the average claim amount
avg_claims_by_type = claims_df_large.groupby('ClaimType')['ClaimAmount'].mean().reset_index()

# Plot the results
plt.figure(figsize=(10, 6))
sns.barplot(x='ClaimType', y='ClaimAmount', data=avg_claims_by_type)
plt.title('Average Claim Amount by Claim Type')
plt.xlabel('Claim Type')
plt.ylabel('Average Claim Amount')
plt.show()
```
